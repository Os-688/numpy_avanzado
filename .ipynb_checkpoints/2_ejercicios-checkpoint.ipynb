```python
import numpy as np
```

## Primer_actividad:

```python
a = np.array([1, 2, 3, 4])

bol = np.where(a == 0)

print(f"El tamaño de las coincidencias con 0 = {len(a[bol])}")
```

## Segunda_actividad:

```python
a = np.array([0, 0, 2, 0])

bol = np.where(a != 0)

print(f"El tamaño de las coincidencias con elementos distintos de 0 = {len(a[bol])}")
```

## Tercera_actividad:

```python
a = np.array([1, 2, np.nan, np.inf])

bol = np.isnan(a)|np.isinf(a)

print(f"El tamaño de las coincidencias con elementos nan e inf es = {len(a[bol])}")
```

## Cuarta_actividad:

```python
a = np.array([1, 2, np.inf])

bol = np.isinf(a)

print(f"El tamaño de las coincidencias con elementos inf es = {len(a[bol])}")
```

## Quinta_actividad:

```python
a = np.array([1, np.nan, 3])

bol = np.isnan(a)

print(f"El tamaño de las coincidencias con elementos nan es = {len(a[bol])}")
```

## Sexta_actividad:

```python
a = np.array([1, 3, 5]) 
b = np.array([2, 3, 4])

c = a == b

print(f"Elemento coincidente: {a[c]}")
```

```python
a = np.array([1, 3, 5]) 
b = np.array([2, 3, 4])

c = a >= b

print(f"Elementos de a mayor o igual que elementos de b: {a[c]}")
```

```python
a = np.array([1, 3, 5]) 
b = np.array([2, 3, 4])

c = a > b

print(f"Elementos de a mayor que elementos de b: {a[c]}")
```

```python
a = np.array([1, 3, 5]) 
b = np.array([2, 3, 4])

c = a <= b

print(f"Elementos de a menor o igual que elementos de b: {a[c]}")
```

```python
a = np.array([1, 3, 5]) 
b = np.array([2, 3, 4])

c = a < b

print(f"Elementos de a menor que elementos de b: {a[c]}")
```

## Septima actividad:

```python
a = np.array([1, 7, 13, 105])
b = a.nbytes
print(f"{b}")
print(f"Tamaño total en memoria (.nbytes): {a.nbytes} bytes")
```
